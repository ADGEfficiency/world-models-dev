```python
import numpy as np
data = []
with open('./rewards.log') as f:
    lines=f.readlines()
    for line in lines:
        data.append(np.fromstring(line, dtype=float, sep=','))
```

```python
data = np.array(data).flatten()
```

```python
import matplotlib.pyplot as plt
f, a = plt.subplots()
%matplotlib inline

plt.plot(data)
```
